```python
#!pip install pandas
import pandas as pd
```

```python
df = pd.read_csv(r'https://stepik.org/media/attachments/lesson/313078/dataset_file_storage.csv', sep=';')
df
```

```python
dm = df.groupby(['ProjectID']).sum()['FileSize'].mean()
dp = df.groupby(['CompanyID','ProjectID']).sum()['FileSize']
dmax = dp.groupby(['CompanyID']).max()
sum(dmax > dm)
```
