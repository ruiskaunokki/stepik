```python
#!pip install pandas
import pandas as pd
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset_345422_14.txt')
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df = df.drop(columns=['Brébeuf (données non disponibles)', 'St-Urbain (données non disponibles)'])
df
```

```python
df.insert(1, 'Weekday', -1)
df
```

```python
for i in range(len(df)):
  df.loc[i, 'Weekday'] = df.loc[i, 'Date'].weekday()
df
```

```python
df['Visiters'] = df[['Berri 1', 'Côte-Sainte-Catherine', 'Maisonneuve 1', 'Maisonneuve 2', 'du Parc', 'Pierre-Dupuy', 'Rachel1']].sum(axis=1)
df
```

```python
stat = df.groupby(['Weekday'])[['Visiters']].mean()
n = 0.0
a = -1
for i in range(len(stat)):
  if n < stat.loc[i, 'Visiters']:
    n = stat.loc[i, 'Visiters']
    a = i
days[a]
```
